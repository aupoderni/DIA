```python
import pandas as pd
import re
```

Reading precolator results with q-value less than 0.005

```python
df_all = pd.read_csv('C:/Users/poder/crux-demo/crux-output/percolator-output/percolator.target.psms.txt', sep = '\t', skipinitialspace=True)
df_prec_mass = df_all.loc[(df_all['percolator q-value'] < 0.005)]
df_percolator = df_prec_mass[['scan', 'sequence', 'charge', 'percolator q-value']]
df_percolator
```

```python
df_all.loc[(df_all['percolator q-value'] < 0.005) & (df_all['spectrum precursor m/z'] == 410.4365)]
```

```python
df_features = pd.read_csv('C:/Users/poder/crux-demo/crux-output/diameter.features.pin', sep = '\t', usecols = ['SpecId', 'ScanNr'])
df_features
```

```python
df_union = pd.merge(df_features, df_percolator, left_on = 'ScanNr', right_on = 'scan')
df_union.drop(['ScanNr', 'scan'], inplace=True, axis=1)
df_union
```

```python
pattern = re.compile(r'target_\d_(.*?)_\d_\d')
for ind in df_union.index:
    df_union.loc[ind,'SpecId'] = re.findall(pattern, df_union['SpecId'][ind])[0]
```

```python
df_union
```

```python
df_union = df_union.rename(columns = {"SpecId": "scan"})
df_union = df_union[['scan', 'sequence', 'charge']]
```
