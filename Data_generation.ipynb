```python
import pandas as pd
```

Reading precolator results with q-value less than 0.006

```python
df_all = pd.read_csv('crux-output/percolator-output/percolator.target.psms.txt', sep = '\t', skipinitialspace=True)
df_prec_mass = df_all.loc[(df_all['percolator q-value'] < 0.006)]
df_percolator = df_prec_mass[['scan', 'sequence']]
df_percolator
```

Replacing the index from df_precolator with the spectrum scan_id

```python
df_features = pd.read_csv('crux-output/diameter.features.pin', sep = '\t', usecols = ['SpecId', 'ScanNr'])
df_features.head(3)
```

```python
df_union = pd.merge(df_features, df_percolator, left_on = 'ScanNr', right_on = 'scan')
df_union.drop(['ScanNr', 'scan'], inplace=True, axis=1)
df_union.head(3)
```

```python
for ind in df_union.index:
    df_union['SpecId'][ind] = df_union['SpecId'][ind].removeprefix('target_0_')
    df_union['SpecId'][ind] = df_union['SpecId'][ind][0:-4]
```

```python
df_union = df_union.rename(columns = {"SpecId": "scan"})
```

```python
df_union
```

Reading diameter results

```python
df_all = pd.read_csv('crux-output/diameter.psm-features.filtered.txt', sep = '\t', skipinitialspace=True, low_memory=False)
df_prec_mass = df_all.loc[(df_all['spectrum precursor m/z'] == 410.4365)]
df_diameter = df_prec_mass[['scan', 'sequence']]
df_diameter
```

Saving data in .csv files

```python
df_diameter.to_csv('train_data_diameter.csv', index=False)
df_union.to_csv('train_data_percolator.csv', index=False)
```
